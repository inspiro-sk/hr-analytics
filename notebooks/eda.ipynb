```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("general_data.csv", index_col='EmployeeID')
df.head()
```

### Clean up the data

```python
df.isna().sum()
```

```python
df.dropna()
df.isna().sum()
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(15, 15))
plt.title('KNN feature space plot', fontsize=14)
plt.xlabel('Monthly Income')
plt.ylabel('Total Working Years')
sns.scatterplot(x=df['MonthlyIncome'], y=df['TotalWorkingYears'], hue=df['Attrition'])
plt.show()
```

### EDA

```python
df.EmployeeCount.value_counts()
```

```python
df.StandardHours.value_counts()
```

```python
# these cols have no predictive power so we can drop them
df.drop(['EmployeeCount', 'StandardHours'], inplace=True, axis=1)
```

```python
df.shape
```

```python
num_cols = list(df.dtypes[df.dtypes != 'object'].index.values)
cat_cols = list(df.dtypes[df.dtypes == 'object'].index.values)
```

```python
num_cols
```

```python
cat_cols
```

```python
# convert attrition to numeric field
df_numeric = df.loc[:, num_cols]
df['Attrition'] = df['Attrition'].astype('category')
df_numeric['Attrition_num'] = df['Attrition'].cat.codes
df_numeric
```

```python
# check for correlations
df_numeric.corr()['Attrition_num'].sort_values()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), cmap='coolwarm')
plt.show()
```
